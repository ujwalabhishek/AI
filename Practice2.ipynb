Q. Select the option that contains the operation that helps you determine which car body type mostly has two doors in the df_cars dataframe grouped by the type of car body.

```python
df_cars.groupby('carbody')['doornumber'].count()
```

```python
df_cars.groupby('carbody')['doornumber'].count().idxmax()
```

```python
df_cars.groupby('carbody')['doornumber'].apply(len)
```

What `df_cars.groupby('carbody')['doornumber']` does:

It groups all the values of 'doornumber' for each carbody.

You can think of it like this:

    {
      'convertible': ['two', 'two', ...],   # 6 items
      'hardtop': ['two', 'four', ...],      # 7 items
      'hatchback': [...],                   # 46 items
      'sedan': [...],                       # 68 items
      'wagon': [...],                       # 20 items
    }

Each group is like a list of values for 'doornumber' for that carbody.

**.apply(len) :** It applies Python's len() function to each group (which is a list-like object). So it counts everything, including nulls.

    ['two', 'two', None, 'four']

.apply(len) will return 4.

**Visual Example:**
Let’s create a small simulated dataset to illustrate the difference:

    df_small = pd.DataFrame({
        'carbody': ['sedan', 'sedan', 'sedan', 'hatchback', 'hatchback'],
        'doornumber': ['four', None, 'two', 'two', None]
    })

Grouped Data:

    sedan:
      doornumber: ['four', None, 'two']
    
    hatchback:
      doornumber: ['two', None]

groupby('carbody')['doornumber'].count():    
    
    sedan       → 2   # (ignores None)
    hatchback   → 1   # (ignores None)

groupby('carbody')['doornumber'].apply(len):

    sedan       → 3   # includes None
    hatchback   → 2   # includes None

✅ Summary:
Operation	            Includes Nulls?	ReturnS

groupby(...).count()	❌ No	        Number of non-null items per group

groupby(...).apply(len)	✅ Yes	        Total number of items per group

Q. How can you summarise the dataframe df_cars by grouping it based on the type of car body and calculating the median price and mean horsepower for each group?

```python
df_cars.groupby('carbody').aggregate({'price': 'median', 'horsepower': 'mean'})
```
