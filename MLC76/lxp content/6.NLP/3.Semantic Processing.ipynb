## Knowledge Graphs


Let’s get started with the first method of semantic processing, which is the knowledge graph.

A graph data structure consists of a finite set of vertices (also called nodes or points), where some pairs are connected through edges.

<div class="MuiBox-root css-1bi8ut6"><div class="text_component" data-testid="text-component"><p dir="ltr">In the case of semantic processing, the text data is aggregated on the graph-based data structure. In the knowledge graph, entities are represented by nodes, where some pairs of the entities are related in some sense. These relations are represented by edges.</p><p dir="ltr">&nbsp;</p><p dir="ltr" style="text-align: center;"><img data-width="497" data-height="210" height="210" width="497" src="https://images.upgrad.com/a7b50d17-7730-408b-8b0a-896a71d2ba4f-Semantic_image_1.png"></p><p dir="ltr">&nbsp;</p><p dir="ltr">These graphs contain the semantic information of different entities. An example of a knowledge graph is given below.</p><p dir="ltr">&nbsp;</p><p style="text-align: center;"><img data-width="753" data-height="286" height="227.88844621513945" width="600" src="https://images.upgrad.com/0bd7a3be-6ba5-4754-a8e8-96e0cfea9d14-Semantic_image_2.png"></p><p>&nbsp;</p><p>You can see the different types of entities, which are human, occupation, names of people and names of places.</p><p dir="ltr">&nbsp;</p><p dir="ltr">The different types of relationships are ‘is a’, ‘was born in’ and ‘is capital of’.&nbsp;</p><p dir="ltr">This is a simple representation of graphs however, graphs are generally very complex structures.&nbsp;</p></div></div>




<div class="MuiBox-root css-1bi8ut6"><div class="text_component" data-testid="text-component"><p dir="ltr">The different types of knowledge graphs are as follows:</p><ul><li dir="ltr" aria-level="1"><p dir="ltr" role="presentation"><a target="_blank" href="https://wordnet.princeton.edu/">WordNet:</a> This is a lexical database of semantic relations between words. It is developed by Princeton University.</p></li><li dir="ltr" aria-level="1"><p dir="ltr" role="presentation"><a target="_blank" href="https://conceptnet.io/">ConceptNet</a>: This is a freely available semantic network that is designed to help computers understand the meanings of words that people use. It is developed by MIT.</p></li></ul><p dir="ltr">The graph that describes the conceptnet is given below.</p><p dir="ltr">&nbsp;</p><p dir="ltr" style="text-align: center;"><img data-width="683" data-height="767" height="600" width="534.2894393741851" src="https://images.upgrad.com/7679852f-c065-497a-b5c6-3308c0d399b2-Semantic_image_3.png"></p><p dir="ltr">&nbsp;</p><p dir="ltr">Both WordNet and ConceptNet are used for natural language understanding. At the end of this session, we will use WordNet to solve a use of Word Sense Disambiguation.</p><p>&nbsp;</p><p dir="ltr">Another type of knowledge graph is UMLS.&nbsp;</p><ul><li dir="ltr" aria-level="1"><p dir="ltr" role="presentation">Unified Medical Language System (UMLS): It is a set of files and software that brings together many health and biomedical vocabularies and standards to enable interoperability between computer systems.</p></li></ul><p dir="ltr">Suppose you need to understand the text data that is related to the medical field. If you use WordNet or ConceptNet, you will have words that do not have relevance, and the results will be inaccurate. Hence, you require domain-specific knowledge graphs.</p><p dir="ltr">We know the famous knowledge graph called Google Search.</p><p dir="ltr">Although these are openly available knowledge graphs, many companies create their own knowledge graphs according to their company requirements.</p><ul><li dir="ltr" aria-level="1"><p dir="ltr" role="presentation">Microsoft uses knowledge graphs for the Bing search engine, LinkedIn data and academics.</p></li><li dir="ltr" aria-level="1"><p dir="ltr" role="presentation">Facebook develops connections between people, events and ideas, focusing mainly on news, people and events related to the social network.</p></li><li dir="ltr" aria-level="1"><p dir="ltr" role="presentation">IBM provides a framework for other companies and/or industries to develop internal knowledge graphs.</p></li><li dir="ltr" aria-level="1"><p dir="ltr" role="presentation">eBay is currently developing a knowledge graph that functions to provide connections between users and the products present on the website.</p></li></ul><p>&nbsp;</p><p dir="ltr">Fun Exploration:</p><p dir="ltr">Watson is a question answering computer system that can answer questions posed in natural language. It is based on knowledge graphs. You can watch Watson winning the famous game of jeopardy <a target="_blank" href="https://www.youtube.com/watch?v=P18EdAKuC1U">here</a>.&nbsp;</p><p>&nbsp;</p><p dir="ltr">Now that you have gone through different types of knowledge graphs, in the next segment, you will gain a detailed understanding of one of the widely used knowledge graphs: ‘WordNet’.</p></div></div>

## Relation Between Senses
### Wordnet

<div class="MuiBox-root css-1bi8ut6"><div class="text_component" data-testid="text-component"><p><strong>Relation Between Senses - Part 1</strong><br>&nbsp;</p><p dir="ltr">In the previous segment, you learnt about different types of knowledge graphs.</p><p dir="ltr">In this segment, you will gain a detailed understanding of the WordNet knowledge graph.&nbsp;</p><p dir="ltr">WordNet is a part of NLTK, and you will use it later in this module to identify the 'correct' sense of a word (i.e., for word sense disambiguation).</p><p dir="ltr">WordNet® is a large lexical database of English words developed by Princeton University. It can be accessed <a target="_blank" href="http://wordnet.princeton.edu/">here</a>.</p><p dir="ltr">Let’s see Jaidev explore the Wordnet database.</p></div></div>

<div class="MuiBox-root css-1bi8ut6"><div class="text_component" data-testid="text-component"><p>Let us understand what we learnt in the above video.</p><p>The diagram below is&nbsp;a&nbsp;screenshot from the WordNet website:</p><p style="text-align: center;"><img data-width="1320" data-height="1548" height="600" width="511.6279069767442" src="https://images.upgrad.com/4890e97a-750b-4331-a91b-1ce3c7f22db1-Untitled (3).png"></p><p>&nbsp;</p><p dir="ltr">In the diagram given above, each word sense of the word bank is grouped into its nouns and verbs. A set of all these senses is called a synset.</p><p dir="ltr">Each sense of the word has a gloss or meaning of the word and an example as in the dictionary.</p><p dir="ltr">For example, the first verb sense of the word has a gloss or a meaning as ‘tip literally’ and the example sentence as ‘the pilot had to bank the aircraft’.</p><p dir="ltr">Similarly, each word sense contains a gloss and an example sentence.</p><p><br>If meanings are available in the dictionary as well, what makes WordNet unique?</p><p dir="ltr">Each of these senses of the words is related to other senses through some relations.&nbsp;</p></div></div>

Now let’s take a look at the relations in WordNet. 

<div class="text_component" data-testid="text-component"><p dir="ltr">The types of relationship between different words can be grouped as follows:</p><ol><li dir="ltr" aria-level="1"><p dir="ltr" role="presentation">Synonym: A relation between two similar concepts</p></li></ol><p dir="ltr">Example: Large is a synonym of big.</p><ol start="2"><li dir="ltr" aria-level="1"><p dir="ltr" role="presentation">Antonym: A relation between two opposite concepts</p></li></ol><p dir="ltr">Example: Small is an antonym of big.</p><ol start="3"><li dir="ltr" aria-level="1"><p dir="ltr" role="presentation">Hypernym: A relation between a concept and its superordinate</p></li></ol><p dir="ltr">&nbsp;A superordinate is all-encompassing.</p><p dir="ltr">Example: Fruits is the hypernym of mango.</p><ol start="4"><li dir="ltr" aria-level="1"><p dir="ltr" role="presentation">Hyponym: A relation between a concept and its subordinate</p></li></ol><p dir="ltr">Example: Apple is the hyponym of fruits.</p><p dir="ltr">You can refer to the diagram given below to understand hyponyms and hypernyms. Any word that is connected with its hypernyms has an ‘is a’ relationship</p><p dir="ltr">&nbsp;</p><p dir="ltr">&nbsp;</p><p style="text-align: center;"><img data-width="557" data-height="144" height="144" width="557" src="https://images.upgrad.com/7d9146ed-db32-4fb5-a8f8-8ce13602cf53-Untitled (4).png"></p><p>&nbsp;</p><ol><li dir="ltr" aria-level="1"><p dir="ltr" role="presentation">Holonym: A relation between a whole and its parts</p></li></ol><p dir="ltr">Example: Face is the holonym of eyes.</p><ol start="2"><li dir="ltr" aria-level="1"><p dir="ltr" role="presentation">Meronym: A relation between a part and its whole.</p></li></ol><p dir="ltr">Example: Eyes is the meronym of human body</p><p dir="ltr">You can refer to the diagram given below to gain an understanding of holonyms and meronyms. Any word is connected with its holonym by a ‘has part’ relationship.&nbsp;</p><p dir="ltr">&nbsp;</p><p dir="ltr" style="text-align: center;"><img data-width="654" data-height="190" height="174.3119266055046" width="600" src="https://images.upgrad.com/1d20ddda-df5f-4cb3-a0b0-6f8e36c796fa-Semantic_image_4.png"></p><p>&nbsp;</p><p dir="ltr">Based on your learnings so far, attempt the following questions.</p><p dir="ltr">&nbsp;</p><p dir="ltr">Apart from these, can you think of some other examples of hypernyms, hyponyms, meronyms, holonyms, synonyms and antonyms?</p><p dir="ltr">To summarise, Wordnet contains word senses for each word, and these senses are related through different relations.</p>
