The basic lexical processing techniques would not be able to identify the difference between the three sentences. Therefore, more sophisticated syntactic processing techniques are required to understand the relationship between individual words in a sentence.</p><p>&nbsp;</p><p>Now, before understanding what syntactic processing is, let’s first understand what syntax is and then take a look at some of the interesting examples that make use of syntactic processing. ## Parts Of Speech

<div class="MuiBox-root css-1bi8ut6"><div class="text_component" data-testid="text-component"><p>Let’s start with the first level of syntactic analysis: PoS (Parts of Speech) tagging. To understand PoS tagging, you first need to understand what parts of speech are.&nbsp;</p><p><br>Let’s consider a simple example given below.</p><p><br>‘You are learning NLP at upGrad.’</p><p><br>From your knowledge of the English language, you are aware that in this sentence, ‘<strong>NLP</strong>’ and ‘<strong>upGrad</strong>’ are nouns, the word ‘<strong>learning</strong>’ is a verb and ‘<strong>You</strong>’ is a pronoun. These are called the parts of speech tags of the respective words in the sentence. A word can be tagged as a noun, verb, adjective, adverb, preposition, etc., depending upon its role in the sentence. These tags are called the PoS tags.&nbsp;</p><p><br>Assigning the correct tag, such as noun, verb and adjective, is one of the most fundamental tasks in syntactic analysis.</p><p><br>Suppose you ask your smart home device the following question.&nbsp;</p><p><br>‘Ok Google, where can I get the permit to work in Australia?’&nbsp;</p><p><br>The word '<strong>permit</strong>' can potentially have two PoS tags: noun or a verb.&nbsp;</p><p><br>In the phrase 'I need a work permit', the correct tag of '<strong>permit</strong>' is 'noun'.&nbsp;</p><p><br>On the other hand, in the phrase<em> </em>'Please permit me to take the exam', the word '<strong>permit</strong>' is a 'verb'.</p><p><br>In the following video, let’s hear from Sumit about part-of-speech tags in detail.</p></div></div>



<div class="text_component" data-testid="text-component"><p>Parts of speech (PoS) are the groups or classes of words that have similar grammatical properties and play similar roles in a sentence. They are defined based on how they relate to the neighbouring words.</p><p><br>Assigning the correct PoS tags helps us better understand the intended meaning of a phrase or a sentence and is thus a crucial part of syntactic processing. In fact, all the subsequent parsing techniques (constituency parsing, dependency parsing, etc.) use the part-of-speech tags to parse the sentence.&nbsp;</p><p><br>Now, let’s understand the types of PoS tags available in the English language.</p></div>




<div class="text_component" data-testid="text-component"><p>A PoS tag can be classified in two ways: <strong>open class and closed class</strong>.</p><p><br>Open class refers to tags that are evolving over time and where new words are being added for the PoS tag.</p><ul><li><strong>Open class</strong><ul><li>Noun</li><li>Verb</li><li>Adjective</li><li>Adverb</li><li>Interjection</li></ul></li></ul><p>Some useful examples of open class PoS tags are as follows:</p><ul><li>Name of the person</li><li>Words that can be added or taken from another language such as words taken from the Sanskrit language such as ‘Yoga’ or ‘Karma’&nbsp;</li><li>Words that are nouns but can be used as verbs such as ‘Google’</li><li>Words that are formed by a combination of two words such as football, full moon and washing machine</li></ul><p>Closed class refers to tags that are fixed and do not change with time.</p><ul><li><strong>Closed class</strong><ul><li>Prepositions</li><li>Pronouns</li><li>Conjunctions</li><li>Articles</li><li>Determiners</li><li>Numerals</li></ul></li></ul><p>Some examples of closed-class PoS tags are as follows:</p><ul><li>Articles: a, an, the&nbsp;</li><li>Pronouns: you and I&nbsp;</li></ul><p>You can take a look at the universal tagsets used by the spaCy toolkit <a target="_blank" href="https://universaldependencies.org/docs/u/pos/">here</a>.</p><p>&nbsp;</p><p><strong>Note</strong>: <a target="_blank" href="https://spacy.io/">spaCy </a>is an open-source library used for advanced natural language processing, similar to NLTK, which you have used in lexical processing.</p><p>&nbsp;</p><p>In this module on syntactic processing, you are going to use the spaCy library to perform syntactical analysis.</p><p>&nbsp;</p><p>You do not need to remember all the PoS tags. You will pick up most of these tags as you work on the problems in the upcoming segments, but it is important to be aware of all the types of tags.</p><p>&nbsp;</p><p>You can also refer to the alphabetical list of 36 part-of-speech tags used in the <a target="_blank" href="https://www.ling.upenn.edu/courses/Fall_2003/ling001/penn_treebank_pos.html">Penn Treebank Project</a>, which is being used by the spaCy library.</p><p>&nbsp;</p><p>In the next segment, you will learn how PoS tagging is done for a sentence.</p></div>

```python
import spacy
nlp = spacy.load("en_core_web_sm")
doc = nlp("upGrad is teaching Data Science courses to the working professionals.")
for token in doc:
    print(token.text, token.pos_, token.tag_)
```

## Hidden Markov Model 

<div class="text_component" data-testid="text-component"><p>In the previous segment, you learnt about PoS tagging based on the frequency of tags alone, which seems inefficient when words are used in different contexts. So, to improve this model, in this segment, you will learn about the Hidden Markov Model, which performs better and uses the context of the previous word to decide the PoS tag of the current word.</p><p>&nbsp;</p><p>Let’s consider the following example; what does your mind process when you see the blank space at the end of this sentence?</p><p>&nbsp;</p><p>'Rahul is driving to _____.'</p><p>&nbsp;</p><p>Don’t you think that the blank space should be the name of a place?</p><p>&nbsp;</p><p>How do you manage to identify that the blank space would be a name of the place?&nbsp;</p><p>&nbsp;</p><p>Try to analyse your thoughts after reading this statement, when you see the word ‘Rahul’ who is driving to some place and hence, you reach to a conclusion that the blank space should be the name of a place (noun).</p><p>&nbsp;</p><p>This means that you have sequentially looked at the entire sentence and concluded that the blank space should be the name of a place.&nbsp;</p><p>&nbsp;</p><p>Now, what if we build an algorithm that can work sequentially to identify the PoS tags of the words based on the PoS tag of the previous word? In the next video, Sumit will explain sequence labelling techniques, particularly Hidden Markov Models, in detail.</p></div>



<div class="MuiBox-root css-1bi8ut6"><div class="text_component" data-testid="text-component"><p>Sequence labelling is the task of assigning the respective PoS tags of the words in the sentence using the PoS tag of the previous word in the sentence.</p><p>&nbsp;</p><p>Hidden Markov Model can be used to do sequence labelling, which means that it takes input of words in a sequence and assigns the PoS tags to each word based on the PoS tag of the previous word.&nbsp;</p><p>&nbsp;</p><p>In the video example ‘The tortoise won the race’, to decide the PoS tag of ‘race’, the model uses the PoS tag of the previous word, i.e., ‘the’, which is an article.&nbsp;</p><p>&nbsp;</p><p>Now, let’s take a look at the following points and try to understand why Hidden Markov Model is called ‘Hidden’:</p><ul><li>When you observe (read/listen) a sentence, you only observe the words in the sentence and not their PoS tags; thus, PoS tags are hidden.</li><li>You must infer these hidden tags from your observations, and that's why Hidden Markov Model is called Hidden.</li></ul><p>&nbsp;</p><p>There are majorly two assumptions that HMM follows, which are as follows:</p><ul><li>The PoS tag of the next word is dependent only on the PoS tag of the current word.</li><li>The probability of the next word depends on the PoS tag of the next word.&nbsp;</li></ul><p><br>Before learning about HMM, you need to understand the two most important types of matrices, i.e., emission and transition matrices. In the next video, let’s take a look at an example of how you can build these matrices.</p></div></div>





<div class="MuiBox-root css-1bi8ut6"><div class="text_component" data-testid="text-component"><p>To build any machine learning model, you first need to train that model using some training data, and then, you need to use that model to predict the output on the test data.</p><p>&nbsp;</p><p>Here, the train data is the corpus of sentences, and you need to perform some manual tasks to assign the PoS tags to each word in the corpus. Once you manually assign the PoS tags to each word in the training corpus, you create two important matrices using the training dataset, which are as follows:</p><ul><li>Emission matrix</li><li>Transition matrix</li></ul><p>&nbsp;</p><p><strong>Note</strong>: We have used only two PoS tags, i.e., noun and verb, as of now to understand the concept in a simpler manner.</p><p>&nbsp;</p><p><strong>Emission matrix: </strong>This matrix contains all words of the corpus as row labels; the PoS tag is a column header, and the values are the conditional probability values.</p><p>&nbsp;</p><p><strong>Note</strong>: Conditional probability is defined as the probability of occurrence of one event given that some other event has already happened. You will get more idea about this in the following example.</p><p>&nbsp;</p><p>Please note that the table shown in the previous video has&nbsp;mistake in numerical values,&nbsp;as the column wise sum of that table is exceeding the value 1. You can consider the following table instead of video one. Also, please consider the values in the&nbsp;table as a reference only to understand the concept, you only need to&nbsp;keep in mind that the column wise sum of Emission matrix will be 1.</p><p>&nbsp;</p><p style="text-align: center;"><img data-width="289" data-height="360" height="360" width="289" src="https://images.upgrad.com/78962037-6690-4d7b-8a7b-2234b5b25336-Capture.PNG"></p><p>&nbsp;</p><p>For example, in the corpus that has been used in the video provided above, whenever a noun appears in the training corpus, there is an 18% chance that it will be the word ‘judge’. Similarly, whenever a verb appears in the training corpus, there is an 84% chance that it will be the word ‘laugh’.</p><p>&nbsp;</p><p>So, here, 0.18 is the probability of occurrence of the word ‘judge’ given that there will be a noun at that place. In a similar way, 0.10&nbsp;is the probability of occurrence of the word ‘judge’ given that there will be a verb at that place.</p><p>&nbsp;</p><p><strong>Transition matrix</strong>: This matrix contains PoS tags in the column and row headers. Let’s try to understand the conditional probability values that have been given in the following table.</p><p>&nbsp;</p><p>Let’s take a look at the first row of the table; it represents that 0.26 is the probability of occurrence of a noun at the start of the sentence in the training dataset.
