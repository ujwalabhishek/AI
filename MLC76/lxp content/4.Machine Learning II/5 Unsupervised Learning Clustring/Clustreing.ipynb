## Supervised vs Unsupervised Learinng


**Supervised machine learning** algorithms make use of **labelled data** to make predictions.

For example, an email will be classified as spam or ham, or a bank’s customer will be predicted as ‘good’ or ‘bad’. You have a target variable Y which needs to be predicted.


In **unsupervised learning**, you are not interested in prediction because you do not have a target or outcome variable. The objective is to **discover** interesting patterns in the data, e.g. are there any **subgroups or ‘clusters’** among the bank’s customers?

### Practical Example of Clustering - Customer Segmentation

Customer segmentation for targeted marketing is one of the most vital applications of the clustering algorithm. 


**Objective:** Here, as a manager of the online store, you would want to **group the customers** into different clusters, so that you can make a **customised marketing campaign for each of the group**. 

You do not have any label in mind, such as good customer or bad customer. You want to just **look at patterns in customer data** and then try and **find segments**. 

This is where clustering techniques can help you with segmenting the customers. **Clustering techniques use raw data to form clusters based on common factors among various data points**. This is exactly what will also be done in segmentation (human defined grouping criterieas), where various people or products will be grouped together on the basis of similarities and differences between them.

As a manager, you would have to decide what the important business criteria are on which you would want to segregate the customers, this can be difficult at times as a human may miss underlying patterns to define the criterias. So, you would need a method or an algorithm that itself decides which customers to group together based on these criteria.


### Clustering vs Segmentation,

 - **Clustering** is an unsupervised machine learning technique. It automatically groups data points into clusters based on natural similarities or patterns found in the data, **without prior knowledge or predefined criteria by humans**. Clustering algorithms **find hidden patterns and segment data based on those similarities**.

 - **Segmentation**, on the other hand, is a human-driven process where **predefined criteria** are used to manually divide a large dataset or customer base into meaningful groups. It involves selecting attributes such as demographics, behaviors, or locations to form segments according to business needs or hypotheses.

### Summary of Differences

| Aspect            | Clustering                           | Segmentation                         |
|-------------------|------------------------------------|------------------------------------|
| Method            | Machine learning, data-driven      | Human-driven, criteria-based        |
| Basis of Grouping | Automatically finds patterns        | Groups based on preselected rules   |
| Purpose           | Explore complex similarities       | Target known or hypothesized groups |
| Flexibility       | Continuously updated and refined   | Static, based on selected attributes|
| Example           | Group customers by behavior patterns| Segment customers by age, gender, location |

Clustering can be used to support and expand segmentation by revealing patterns that humans may not notice. Both are important for targeting and understanding groups well but come from different starting points: clustering from data exploration, segmentation from business intent.




### Steps:

 - Group customers into clusters based on features and then analyse the groups to derive actionable insights
 - Segmentation is done to do the clustering
 - Segmentation needs to be stable - people segnented in a certian segment need to be in same segment and not fluctate to other segment. This means that the same person should not fall under different segments upon segmenting the data on the same criteria.
 - Segments should have intra-segment homogeneity and inter-segment heterogeneity. You will see in later sessions how this can be defined mathematically.



### Types of Segmentation for Customer 

 - **Behavioural segmentation:** Segmentation is based on the actual patterns displayed by the consumer. eg youtube rowsing pattern , what kind of contenet, time of login
 - **Attitudinal segmentation:** Segmentation is based on the beliefs or the intents of people, which may not translate into similar action. eg when going for shopping only choosing randed stuff , what brands, brand loyality
- **Demographic segmentation:** Segmentation is based on a person’s profile and uses information such as age, gender, residence locality, income, etc. eg income groups living in certin areas of the City

### Euclidean Distance

We have learnt that clustering works on the basis of grouping the observations which are the most similar to each other. What does this exactly mean?

 

In simple terms, the algorithm needs to find data points whose values are **similar** to each other and therefore these points would then belong to the same cluster. The method in which any clustering algorithm goes about doing that is through the method of finding something called a **“distance measure”**. The distance measure that is used in K-means clustering is called the **Euclidean Distance** measure. Let’s look at the following lecture to understand how this value is calculated.

The idea of distance measure is quite intuitive. Essentially, the observations which are closer or more similar to each other would have a low Euclidean distance and the observations which are farther or less similar to each other would have a higher Euclidean distance. 



### Centroid

centroids are essentially the **centre points of triangles**. Similarly, in the case of clustering, centroids are the centre points of the clusters that are being formed.

**Intution:** Imagine you have the following clusters of the marks of a group of students in Mathematics and Biology and someone asks you to explain them. From a glance, you can easily interpret the 4 clusters that are being formed. 



So the four clusters that are being formed are as follows:

- Cluster 1: Students who have scored high marks in Bio, but poor marks in Maths
- Cluster 2: Students who have scored average  marks in Bio and  Maths
- Cluster 3: Students who have scored high marks in both Bio and Maths
- Cluster 4: Students who have scored high marks in Maths, but poor marks in Bio

 
Now the above representation is fine and correct, but it is **missing one crucial information** - the **numerical order**. For example, when you want to compare two clusters say Cluster 1 and Cluster 2 can you say by how much marks on average do the students from Cluster 1 outperform or underperform the Cluster 2 students in a particular subject just by taking a look at the above visualisation alone? Is it by 10 marks? Or 15?

This is where the concept of **Centroids** come in handy.
