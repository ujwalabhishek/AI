## Vision Transformers - Theory



<div class="MuiBox-root css-1bi8ut6"><div class="text_component" data-testid="text-component"><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">As explained in the video, the transformer module is a fundamental component in language modelling and is extensively used in various natural language processing (NLP) tasks.</p><p style="text-align: justify;">&nbsp;</p><p><img style="max-width: 100%; display: block; margin-left: auto; margin-right: auto; cursor: pointer; border: 2px solid transparent; text-align: justify;" data-image-url="https://d35ev2v1xsdze0.cloudfront.net/de73ba88-8e37-4ff9-9edb-cc44a1d0505b-w3rmzbwj.png" class="image-center" src="https://d35ev2v1xsdze0.cloudfront.net/de73ba88-8e37-4ff9-9edb-cc44a1d0505b-w3rmzbwj.png"></p><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">The transformer module comprises several key layers, including the following:</p><p style="text-align: justify;">1. <strong>Self-attention laye</strong>r: This layer helps in capturing the relationship between different words (tokens) in a sentence.<br>2. <strong>Normalisation layer:</strong> This layer stabilises the learning process and improves convergence.<br>3. <strong>Feedforward layer:</strong> This layer processes the normalised output further to produce the final embeddings.</p><p style="text-align: justify;">The input to the transformer module comprises word tokens that are first converted into word embeddings. These embeddings are then enriched with positional encodings to retain the order of words in a sentence, which are then fed into the transformer encoder.<br>&nbsp;</p><p style="text-align: justify;">After completing the Transformer module, the professor explained the self-attention mechanism.</p><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">The self-attention mechanism is central to the transformer's functioning. It involves three learnable matrices that transform the input word tokens into the following three distinct linear embeddings:</p><p style="text-align: justify;">1. Query embeddings (Q)<br>2. Key embeddings (K)<br>3. Value embeddings (V)</p><p style="text-align: justify;">The goal is to derive an enhanced representation of the input tokens. This is achieved by computing a weighted sum of the value vectors, where the weights are determined by the dot product of the query and key embeddings.&nbsp;</p><p style="text-align: justify;">&nbsp;</p><p><img style="max-width: 100%; display: block; margin-left: auto; margin-right: auto; cursor: pointer; border: 2px solid transparent; text-align: justify;" data-image-url="https://d35ev2v1xsdze0.cloudfront.net/2e109a92-11e1-490d-ae39-795aaae64341-6ke4j9ud.png" class="image-center" src="https://d35ev2v1xsdze0.cloudfront.net/2e109a92-11e1-490d-ae39-795aaae64341-6ke4j9ud.png"></p><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">The self-attention process is summarised below:</p><ul><li style="text-align: justify;"><strong>Input:</strong> Start with n tokens, each represented as a vector of dimension d.</li><li style="text-align: justify;"><strong>Transformation: </strong>These tokens are converted into three sets of embeddings: n query tokens and n key embeddings, both with dimension dq, and n value embeddings with dimension dv.</li><li style="text-align: justify;"><strong>Output: </strong>Each token is transformed into a more meaningful representation, which is a weighted sum of the value embeddings. The weights are calculated using matrix A, which is derived from the dot product between the query and key embeddings.</li></ul><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">Then, he explained the normalisation technique in transformers, specifically layer normalization.&nbsp;</p><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">Unlike batch normalization, layer normalization is used in transformers. To understand the reason for this, they have been compared in this table.</p><p style="text-align: justify;">&nbsp;</p><p><img style="max-width: 100%; display: block; margin-left: auto; margin-right: auto; cursor: pointer; border: 2px solid transparent; text-align: justify;" data-image-url="https://d35ev2v1xsdze0.cloudfront.net/0c7845a5-0b25-47de-aee9-c32881c660ac-g2ys3y0m.png" class="image-center" src="https://d35ev2v1xsdze0.cloudfront.net/0c7845a5-0b25-47de-aee9-c32881c660ac-g2ys3y0m.png"></p><p style="text-align: justify;">&nbsp;</p><table border="1" cellspacing="1" cellpadding="1"><tbody><tr><td style="text-align: justify;"><strong>Batch Normalization</strong></td><td style="text-align: justify;"><strong>Layer Normalization</strong></td></tr><tr><td style="text-align: justify;">Normalises across each feature dimension within a batch</td><td style="text-align: justify;">Normalises across each feature dimension for individual samples independently</td></tr><tr><td style="text-align: justify;">Considers the batch dimension</td><td style="text-align: justify;">Does not consider the batch dimension</td></tr><tr><td style="text-align: justify;">Designed for fixed-length input sequences within a batch</td><td style="text-align: justify;">Accommodates variable-length input sequences within a batch</td></tr><tr><td style="text-align: justify;">The mean and standard deviation are computed for each feature dimension by averaging across samples in the mini-batch</td><td style="text-align: justify;">The mean and standard deviation are computed for each feature dimension within a single sample</td></tr></tbody></table><p style="text-align: justify;"><br>In conclusion, layer normalization is preferred in transformers because it effectively handles variable-length inputs within a single batch.</p><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">Finally, the professor explained the concept of CLS tokens.</p><p style="text-align: justify;">&nbsp;</p><p><img style="max-width: 100%; display: block; margin-left: auto; margin-right: auto; cursor: pointer; border: 2px solid transparent; text-align: justify;" data-image-url="https://d35ev2v1xsdze0.cloudfront.net/cca25ec5-cafa-4c85-8ea5-ebcc0a4cdc1f-71y3aoti.png" class="image-center" src="https://d35ev2v1xsdze0.cloudfront.net/cca25ec5-cafa-4c85-8ea5-ebcc0a4cdc1f-71y3aoti.png"></p><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">'Classification (CLS) tokens are special tokens added to the beginning of an input sentence in transformers. A CLS token serves to generate a generic sentence-level representation by leveraging global attention across all word embeddings in the sentence. An MLP can use the output embedding of the CLS token to make decisions about the sentence as a whole. This token helps mitigate biases present in individual word embeddings, providing a neutral and comprehensive representation for tasks such as classification.</p><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">In the next segment, you will learn about each of the individual components of the ViT architecture.</p><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">Before you proceed to the next segment, please attempt the question given below.</p></div></div>

## Architecture of Vision Transformers 

<div class="MuiBox-root css-1bi8ut6"><div class="text_component" data-testid="text-component"><p style="text-align: justify;">ViTs are emerging as a powerful alternative to convolutional neural networks (CNNs) for image recognition tasks. The ViT architecture leverages a transformer encoder, similar to those in language processing, comprising multi-head attention, normalisation and MLP layers.&nbsp;</p><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">In the next video, Professor Viswanath will explain the architecture of ViT in detail.</p></div></div>



<div class="MuiBox-root css-1bi8ut6"><div class="text_component" data-testid="text-component"><p>&nbsp;</p><p>As explained in the video, the architecture of a ViT is notably similar to the architecture of the transformer encoder used in language processing. The core components of the ViT encoder include multi-head attention, a normalisation layer and an MLP.&nbsp;</p><p>&nbsp;</p><p><img style="max-width: 100%; display: block; margin-left: auto; margin-right: auto; cursor: pointer; border: 2px solid transparent;" data-image-url="https://d35ev2v1xsdze0.cloudfront.net/3c95a59f-5893-4bb2-a4e3-675e9cb984ed-v1oluu6n.png" class="image-center" src="https://d35ev2v1xsdze0.cloudfront.net/3c95a59f-5893-4bb2-a4e3-675e9cb984ed-v1oluu6n.png"></p><p>However, the key difference lies in the method of feeding input to the transformer encoder module. Unlike traditional methods, ViTs require images to be processed in a specific way to fit into the transformer architecture.
