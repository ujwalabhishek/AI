## Evolution of Machine Translation

there are many other languages all over the world. 

With the help of machine translation, we can remove this barrier between different languages and have seamless communication between two people from different parts of the globe.





<div class="MuiBox-root css-1bi8ut6"><div class="text_component" data-testid="text-component"><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">Mapping an input sentence present in one language to a sentence in another language is called machine translation. For a long period of time, translation was a labour-intensive process that depended exclusively on human effort. Though this process is reliable, it takes a lot of time and is computationally expensive too. To accurately translate the given text, translators have to depend on a volumetric amount of information within the required time.&nbsp;</p><p style="text-align: justify;"><br>In 1949, Warren Weaver gave the idea of language translation using machines. To perform translation, all the linguistic relationship between the source and the target language was studied. This kind of translation was called <strong>Rule-Based Machine Translation (RBMT)</strong> as it was based on dictionaries and grammar.</p><p style="text-align: justify;"><br>As years progressed, different methods of translation emerged paving the way for better and quicker translation. Integrating statistical models helped get the job done by finding the statistically most likely translation for use. This kind of approach, known as <strong>Statistical Machine Translation (SMT)</strong>, performed better than RBMT and was used across the 1990s.&nbsp;</p><p style="text-align: justify;">With the development of neural-based language models by Yoshua Bengio, it was understood that the neural networks-based model could perform better in understanding languages. His work laid the foundation for using neural networks in machine translation.</p><p style="text-align: justify;"><br>In 2014, Sutskever et al. and Cho et al. developed a model known as <strong>sequence to sequence (seq2seq) </strong>for the task of <strong>Neural Machine Translation(NMT)</strong>. This model used long short-term memory (LSTM) for both encoder and decoder, which together framed end-to-end learning of the model. As an encoder, LSTM maps the input sequence to a vector of fixed dimensions. This vector is then propagated to the decoder as a context. Using this context vector, another LSTM decodes a target sequence which forms the translation.&nbsp;</p><p style="text-align: justify;"><br>To understand more about what a sequence to sequence model is, let’s move to the next segment.&nbsp;</p></div></div>

**Understanding Sequence to Sequence Model through a Game**

So we've talked about the sequence to sequence model.Let's understand this and get the intuition through a game.So assume that you have a school of participants and there are two classes and they're participating in this game.It's the job of the first class to actually encode a message, and it's the job of the second class to actually decode that message.It could be a translated text.So the first class gets told a language in English, that's the job.The second class to translate in Hindi.So the rules of the game work like this that the the first student actually gets past the first word.What he has to do is encode this word and pass it to the second.The second person gets another word.It's his job to use what has been said in the first by the first student, mix it with the information which he's been told and pass it to the third and so on.And this game goes on till the last person who gets the stop token.So at this point, this person's responsibility is to pass to the next class of context.So we call this as a context.He passes the entire coded whatever he has heard, he passes this to the next class, to the first person.Since these guys have played this game quite a long time, they know how to decode this message.So it's the first person who says mujhe because he knows that that's what the essence of this translation is.What he says gets heard by the second person.He also get this encoded message.Now it's for this second person to start making sense of what's been said to him.He he makes a judgement and says the second word and so on till the last person gets to say that there's a stop token.So this is the game which is essentially what we will be calling the encoder decoder.So in our sequence to sequence model, we call this as the encoder class and we call this as the decoder class.In a sense, this is what is actually happening in the sequence to sequence model.

### Requirements of NMT Architecture

To produce an efficient translation, the architecture should be robust enough to handle the variable length of the input sentence and the output sentence. It should be able to understand the subject, object and action present in the sentence to produce the best translation. Let’s understand what conditions an efficient NMT model is able to satisfy. 



<div class="MuiBox-root css-1bi8ut6"><div class="text_component" data-testid="text-component"><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">A good NMT model should have the following characteristics to produce a good translation for an input sentence:</p><ul><li style="text-align: justify;"><strong>It should handle sequences of variable length</strong>: The output of the NMT model should not be dependent on the input length such that&nbsp;any length input can be passed in and any length output can be generated.&nbsp;</li><li style="text-align: justify;"><strong>It should be able to maintain the correct information</strong>: While providing translation, the decoder should maintain the syntactic as well as the semantic information as stated in the input sequence, else the meaning may change entirely.</li><li style="text-align: justify;"><strong>It should share parameters across the sequences</strong>: To train the entire seq2seq model, all the parameters should be shared such that during the backpropagation, both the encoder and the decoder could be trained simultaneously.&nbsp;</li><li style="text-align: justify;"><strong>It should track long-term dependencies</strong>: It should be able to remember information over long time periods.</li></ul></div></div>

**When we pass long sentences to the RNN, it is not able to remember all the required information**  and tends to forget some parts. In the next video, Mohit will focus on the drawbacks of a vanilla RNN and how LSTM plays a key role in mitigating those drawbacks.




LSTM stands for **Long Short-Term Memory**

It is a specialized type of **Recurrent Neural Network (RNN)** architecture used in deep learning, designed to learn and retain long-term dependencies in sequential data, such as time series, speech, and text. 


**Key Aspects of LSTM**

**Purpose:** LSTMs were created to overcome the "vanishing gradient problem" of traditional RNNs, which prevents them from remembering information over long sequences.

**Structure:** An LSTM unit consists of a memory cell and three gates—**forget gate**, **input gate**, and **output gate** —that regulate the flow of information.

**Applications:** They are widely used for machine translation, speech recognition, time-series forecasting, and video analysis. 

<hr>

**Why "Long Short-Term"?**

**Short-term:** Like RNNs, they process data in a sequence, updating their state at each step.

**Long-term:** The memory cell (or "cell state") allows the network to carry relevant information across many time steps, acting as a long-term memory, while the gates decide what to store or discard.

Teacher forcing is a training technique for sequence-to-sequence neural machine translation (NMT) models that feeds the correct target token (ground truth) from the previous time step as input for the current time step, rather than the model's own, potentially incorrect, predicted output. This accelerates convergence, stabilizes training, and prevents errors from compounding early on

**Key Details About Teacher Forcing in NMT:**

- **Mechanism:** In an RNN/LSTM decoder, instead of feeding the output prediction at time 
 to the input at time , the system uses the actual target sentence word from the training data.

- **Pros:**

    - **Faster Training:** The model learns faster because it is not confused by its own early, inaccurate predictions.
    - **Stability:** Prevents "compounding errors" where an early mistake leads to gibberish, allowing the model to focus on learning grammar and word mappings.

- **Cons & Limitations:**

    - **Exposure Bias:** A significant discrepancy arises between training (using ground truth) and inference/testing (using previous predictions), which can lead to poor, unstable performance when the model is on its own.
    - **Local Optimization:** It lacks global planning, as it focuses only on predicting the immediate next word rather than the overall sentence structure.
      
**Alternatives:** To mitigate exposure bias, methods like scheduled sampling are sometimes used, which gradually mix ground-truth tokens with predicted tokens during training.


<div class="MuiBox-root css-1bi8ut6"><div class="text_component" data-testid="text-component"><p style="text-align: justify;"></p><p style="text-align: justify;">As we have seen earlier, the decoder is trained to predict the next characters of the target sequence given the previous prediction. But what if, for a target sequence “<strong>A boy climbing up the ladder</strong>”, the model at the second timestep produces a wrong prediction – “dog” instead of “boy”? For the third timestep, this will be fed back into the model, and it will lead to another wrong prediction. This process will be continued and at each timestep, the loss will accumulate, and it becomes difficult for the model to learn.<br>Once the entire sequence is predicted, we may get a totally different prediction – "<strong>A dog running on the streets</strong>”.&nbsp;</p><p style="text-align: justify;">Let’s try to understand this using another example. Imagine an exam where you are presented with a question that has multiple parts. Each question is based on the answer to the previous question. So, if you answer a question incorrectly, it will carry forward to the rest of the questions following it, leading to an overall poor score.&nbsp;<br>What if for each question, your teacher checks your response immediately and lets you know the correct answer? With this kind of approach, you will always have a better chance in answering the next question.&nbsp;</p><p style="text-align: justify;">When the same analogy is applied to model training, it is called <strong>teacher forcing</strong>, where the model is fed with the correct target token at each timestep, regardless of the model's predictions at previous timestep. With this technique, the model becomes more robust then when it was sometimes fed its own predictions at the previous timestep.<br>Therefore, the decoder learns to generate target[ : i+1] given target[ : i] and thus helps inconverging to a better model faster.
    
#### Model Inference
<p style="text-align: justify;">During inference i.e during testing, we change the prediction process slightly.</p><ul><li style="text-align: justify;">Once the decoder receives the context vector, it is fed with an input token <strong>&lt;start&gt;.</strong></li><li style="text-align: justify;">Based on this input, the trained decoder will produce a list of predictions (probability scores) for the first word.&nbsp;</li><li style="text-align: justify;">By taking the argmax of the predictions, the next word is computed and appended to an empty list <strong>‘Result’.</strong></li><li style="text-align: justify;">This previous prediction, stored in ‘Result’, is then sent as an input to the model in the next timestep.&nbsp;</li><li style="text-align: justify;">The entire process is repeated until the model produces an &lt;end&gt; token, which indicates the end of the prediction process.&nbsp;</li></ul><p style="text-align: justify;">&nbsp;</p>

#### BLUE Score

Now that you have understood the whole model training process, let’s look into the evaluation metric that we employ for model prediction – the BLEU score.

#### Model Inference
    
<p style="text-align: justify;">Once the model is trained&nbsp;you need to create a&nbsp;<code translate="no" dir="ltr">evaluate</code>&nbsp;function to run the model inference.</p><p style="text-align: justify;">Overall the function is similar to <code translate="no" dir="ltr">train_step</code>&nbsp;except that the input to the decoder at each time step is the last prediction from the decoder.&nbsp;</p><p style="text-align: justify;">The major changes to the function are as follows:</p><ul><li style="text-align: justify;">The input text is processed first by&nbsp;<code translate="no" dir="ltr">preprocess_sentence</code>&nbsp;and padded using the&nbsp;<code translate="no" dir="ltr">keras.preprocessing.sequence.pad_sequences</code>&nbsp;.</li><li style="text-align: justify;">Once the text is converted to their respective token IDs, the trained decoder will produce a list of predictions (probability scores) for the first word.&nbsp;By taking the argmax of the predictions, the next word is computed and appended to an empty list ‘Result’.</li><li style="text-align: justify;">This previous prediction, stored in ‘Result’, is then sent as an input to the model in the next timestep.&nbsp;</li></ul><p style="text-align: justify;">The entire process is repeated until the model produces an &lt;eos&gt; token, which indicates the end of the prediction process.&nbsp;</p><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">NOTE:&nbsp;To avoid the&nbsp;situation&nbsp;when no n-gram overlaps are found, a smoothing function can be used.
