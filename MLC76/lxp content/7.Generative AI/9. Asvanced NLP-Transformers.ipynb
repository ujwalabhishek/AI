# The Birth of Transformers

<div class="MuiBox-root css-1bi8ut6"><div class="text_component" data-testid="text-component"><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">Throughout this program, with each module, you have seen how different algorithms have evolved and rendered the previous generation of algorithms obsolete through their superior performance and better results.<br>In the field of NLP, you would have seen how recurrent neural networks aided in the processing of text with their capabilities to process information through recurrent neurons. However, they suffered from long-term dependencies when longer/multiple sentences were fed to them, as they could not retain information and were thus unable to understand the context of the information fed to them.</p><p style="text-align: justify;"><br>This was an issue, particularly in the encoder-decoder setting, where the information has to travel between two sets of recurrent networks. In the previous module, we also raised an important question: Is the last hidden state enough to capture global information pertaining to neural machine translation?<br>This gap was bridged through the introduction of an attention mechanism, which helped in creating a context vector, that can effectively capture the relative importance of each token in a given sequence.&nbsp;</p><p style="text-align: justify;"><br>Using the cosine similarity between a decoder's hidden state and an encoder’s hidden state, an attention network tells us what states of the encoder are important at a particular time stamp. Therefore, the attention mechanism gives the decoder access to ‘look-back’ at all of the encoder’s hidden states based on its current state. This allows the decoder to extract only relevant information from the encoder states (key) for a particular decoder’s hidden state (query), thus learning more complicated dependencies between the input and the output.<br>But even with the advent of such a mechanism, text processing is still executed as a sequential process. The architecture of transformers was created to resolve this problem. You will understand this in the following video.</p></div></div>



<div class="text_component" data-testid="text-component"><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">As Ankush explained, the design of the <b>transformer architecture is inspired by the combination of relative dependencies provided by attention mechanisms and the parallel processing capability provided by a convolutional neural network (CNN).</b><br>With such an architecture, the model can process each sequence in parallel and give a context vector that can map the relational dependencies of any input fed to it.</p><p style="text-align: justify;"><br>The paper from the Google Brain team – <a target="_blank" href="https://arxiv.org/abs/1706.03762">‘Attention Is All You Need’</a> – was a breakthrough and one of the most revolutionary papers in the field of NLP. It introduced everyone to the architecture of transformers and changed the progress of research in the domain of natural language understanding (NLU). It proposes a novel approach to encode positional information and apply the attention mechanism parallelly, thus accelerating the training.&nbsp;</p><p style="text-align: center;">&nbsp;</p><p><img alt="Transformer Architecture" class="image-center" style="cursor: pointer; border: 2px solid transparent; display: block; margin-left: auto; margin-right: auto; text-align: center;" src="https://images.upgrad.com/76b0fae9-8333-4333-9c7a-fdedeb46a615-Transformers.png"></p><p style="text-align: justify;"><em>“We propose a new simple network architecture, the Transformer, based solely on attention mechanisms, dispensing with recurrence and convolutions entirely”</em>. - Authors(Attention is All you Need).</p><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;"><br>However, the architecture does not look simple at all. Therefore, let’s dissect each block to understand the whole working of a transformer.</p><p style="text-align: justify;">The base architecture of a transformer is like a sequence-to-sequence model consisting of two main building blocks: an encoder and a decoder block. The encoder generates a sequence of embedding vectors 𝒁 = (𝒛₁, …, 𝒛ₙ) from an input representation sequence x = (𝒙₁, …, 𝒙ₙ). Using this contextual information (𝒁), the decoder generates the output sequence y = (𝒚₁, …, 𝒚ₘ), one at a timestep.&nbsp;<br>&nbsp;</p><p style="text-align: justify;"><strong>Note:</strong> Here, the decoder produces the output using an auto-regressive technique, where the model uses the previously generated word along with the context vector to predict the next word.</p><p>&nbsp;</p></div>

### Let’s break this architecture down



<div class="MuiBox-root css-1bi8ut6"><div class="text_component" data-testid="text-component"><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">As explained in the video, a transformer has multiple encoders and decoders stacked on top of each other (stack of N = 6 identical layers).<br>Each layer is composed of sub-layers, whose main components are as follows:</p><ul><li style="text-align: justify;">Multi-head self-attention mechanism</li><li style="text-align: justify;">Position-encoding and position-wise feed-forward NNs</li></ul><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">We will cover each of these layers and components individually to help you understand the complete architecture at the end.<br>In the next segment, you will understand the encoder layer and its components.</p><p style="text-align: justify;">&nbsp;</p><b>Additional Reading:</b><p style="text-align: justify;"><strong>How Uber Predicts Arrival Times Using Transformers</strong></p><p style="text-align: justify;">Check out <a target="_blank" href="https://www.uber.com/en-IN/blog/deepeta-how-uber-predicts-arrival-times/">this </a>blog from Uber to understand how their ML team have used Transformers to predict ETA.</p></div></div>

<p style="text-align: justify;"><strong>Understanding the Encoder</strong></p>
<p style="text-align: justify;">The role of an encoder is to observe an entire input sequence in a single shot and map the relative importance of one token corresponding to other tokens present in the sequence using the principle of the attention mechanism.</p><p style="text-align: justify;"><br>In the next video, Ankush will introduce you to the components inside an encoder block and how they process each input.</p>



<div class="MuiBox-root css-1bi8ut6"><div class="text_component" data-testid="text-component"><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">As explained in the video, any input fed into an encoder undergoes the following steps:</p><ul><li style="text-align: justify;">A given input sequence (“a black dog on blue mat”) is converted into its respective tokens(‘a’, ‘black’, ‘dog’, ‘on’, ‘blue’, ‘mat’).</li><li style="text-align: justify;">Each token is then translated into its respective numerical representations using a tokenizer, which is then passed on to the embedding layer.&nbsp;</li><li style="text-align: justify;">The input embedding layer generates meaningful embeddings of dimension 512 for each token. For example, if the shape of the input tokens is 4*1, then the input embedding layer converts the entire input into the shape of 4*512 (1 vector of size 512 for each token).&nbsp;</li><li style="text-align: justify;">Words of a similar family have similar embeddings. Therefore, if we get the words ‘blue’ and ‘black’, their embeddings should be similar. However, the same would not be the case for the words ‘black’ and ‘on’.</li></ul><p style="text-align: justify;">&nbsp;</p><p style="text-align: center;">&nbsp;</p><p><img class="image-center" style="cursor: pointer; border: 2px solid transparent; display: block; margin-left: auto; margin-right: auto; text-align: center;" src="https://images.upgrad.com/10ee76b3-66f3-4e06-be3f-9baf549e0660-Word_embeddings.png"></p><p style="text-align: justify;">Representation of the process for an example where the dimension of each token is 4</p><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">However, if these inputs are provided in parallel, the information about their positions will be lost. Therefore, the embedding layer will not be able to account for the position at which any two words appear. So, how can we pass the information to the encoder that the words ‘black’ and ‘blue’ are located far from each other in the given sentence, even though they sound similar?&nbsp;</p><ul><li style="text-align: justify;">To resolve this problem, we must pass another embedding, along with the input embedding, to ensure the information about the position of every token is also considered. These embeddings are called positional embeddings.</li></ul><p style="text-align: justify;">&nbsp;</p><p style="text-align: center;">&nbsp;</p><p><img class="image-center" style="cursor: pointer; border: 2px solid transparent; display: block; margin-left: auto; margin-right: auto; text-align: center;" src="https://images.upgrad.com/785d84fa-542a-4c08-a122-f8e6e8e28306-PE.png"></p><p style="text-align: justify;"><i>“Since our model contains no recurrence and no convolution, in order for the model to make use of the order of the sequence, we must inject some information about the relative or absolute position of the tokens in the sequence”. </i>– Authors(Attention is All you Need).</p><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">&nbsp;</p><ul><li style="text-align: justify;">You can refer to these embeddings as a disturbance added to their original embeddings, such that sufficient distances are provided to them in the form of their position.&nbsp;</li><li style="text-align: justify;">However, if we push the words too far away, the semantic information will be irrelevant with respect to the positioning information. Using Fourier analysis, the authors of the Transformer model proposed an index-dependent function that encodes the position of each word as a sinusoidal wave while keeping low values.</li></ul><p style="text-align: center;">&nbsp;</p><p><img class="image-center" style="cursor: pointer; border: 2px solid transparent; display: block; margin-left: auto; margin-right: auto; text-align: center;" src="https://images.upgrad.com/699fab74-c95a-4cd6-b54c-0e92156c4ab0-pushing_vectors.png"></p><p style="text-align: center;">If we push the words too far away, the semantic information will be irrelevant with respect to the positioning information.&nbsp; The diagram above represents the addition of these disturbances or the ‘push’ that the position vectors exert on the embeddings. The arrows in blue push towards one cluster while the arrows in red push towards another cluster.</p></div></div>

### Positional encodings / embeddings in detail



<div class="MuiBox-root css-1bi8ut6"><div class="text_component" data-testid="text-component"><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">The addition of these vectors should satisfy certain conditions for the efficient processing of information:</p><ol><li style="text-align: justify;">It should output a unique encoding for each time step (word’s position in a sentence).</li><li style="text-align: justify;">The distance between any two time steps should be consistent across sentences of different lengths.</li><li style="text-align: justify;">Our model should generalise to longer sentences without any effort. Its values should be bounded.</li><li style="text-align: justify;">It must be deterministic.</li></ol><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">To satisfy all these constraints, the author of the Transformer model suggested the use of the sine and cosine waves of different frequencies.&nbsp;Using Fourier analysis, the authors proposed an&nbsp;index-dependant function&nbsp;that encodes the position of each word as a sinusoidal/cosine wave while keeping low values.</p><p style="text-align: justify;">&nbsp;<img data-latex="PE_{(pos,2i)} = sin(\frac{pos}{10000^{\frac{2i}{d}}})" alt="Equation" class="image-center" style="cursor: pointer; border: 2px solid transparent; display: block; margin-left: auto; margin-right: auto;" src="https://latex.upgrad.com/render?formula=PE_%7B(pos%2C2i%2B1)%7D%20%3D%20cos(%5Cfrac%7Bpos%7D%7B10000%5E%7B%5Cfrac%7B2i%7D%7Bd%7D%7D%7D)"></p><p style="text-align: justify;">Here, pos is the position of the word, i is the i-th dimension of the word embedding, and d/dmodel is the number of dimensions in the embeddings.</p><p style="text-align: justify;"><br>In summary, for every token at index ‘pos’ in the output embedding from the ‘input embedding’ layer, we will generate a position vector of dimension 512 (i.e., d_model that is equal to the embedding dimension for each token), where for every even embedding index (from 0 →512 i.e 0,2,4,6,…..510), we follow the first formula, and for odd indexes (1,3,5,7….511), we follow the second one.</p><p style="text-align: justify;"><br>Having a periodic function ensures values are mapped in a given interval. Also, we can avoid saturation of values as seen using sigmoid or hyperbolic tangent functions. Each position is completely identified by the frequency and offset of the sine/cosine wave.</p><p style="text-align: center;">&nbsp;</p><p><img data-width="2299" data-height="1920" height="501.0874293170944" width="600" class="image-center" style="cursor: pointer; border: 2px solid transparent; display: block; margin-left: auto; margin-right: auto; text-align: center;" src="https://images.upgrad.com/a657efb8-4e81-4f83-9938-3d22b44c6933-wave.png"></p><p style="text-align: justify;">Distinct positional encodings are created for each dimension and position</p>
<p style="text-align: justify;">Therefore, by adding a positional encoding block before passing the embedded output to the encoder, we can successfully pass the positional information of the respective tokens.</p><p style="text-align: justify;">Once the positional embeddings are added, next comes the core of the encoder - ‘Attention’. In the next segment, we take a deep dive into the self-attention mechanism in transformer architecture.</p><p style="text-align: justify;">Before you proceed to the next segment, please attempt the question given below.</p></div></div>

# Self-Attention in a Transformer



<p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">Self-attention, sometimes called intra-attention, is an attention mechanism to compute relative dependencies of the given sequence. As the transformer model processes each word (each position in the input sequence), self-attention allows it to look at other positions in the input sequence for clues that can help lead to better encoding of that word.</p><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">Let's dive right in and understand the application of self-attention in transformer architecture in the next video.</p>

* **Value (V)** = *What information I carry*

# 🔹 Real-life analogy (very powerful)

Think of **Google Search**:

* You type a search → **Query**
* Web pages have titles/keywords → **Keys**
* Actual content of pages → **Values**

👉 Google matches:

```id="y3q1t6"
Query ↔ Key → find relevance
```

Then returns:

```id="g1x8mv"
Value (actual content)
```

---

# 🔹 Map this to words in a sentence

Sentence:
👉 "The cat sat on the mat"

Pick word: **"sat"**

---

## Step 1: Each word creates 3 vectors

For every word:

* Q → what this word is looking for
* K → what this word represents
* V → actual information of the word

So for "cat":

* Q_cat, K_cat, V_cat

For "sat":

* Q_sat, K_sat, V_sat

---

## Step 2: Matching (this is the magic)

👉 "sat" asks:

> "Which words are important for me?"

So:

```id="4wq8v1"
Q_sat compared with K_cat, K_mat, K_the...
```

This gives **attention scores**

Example:

* Q_sat ⋅ K_cat → high ✅
* Q_sat ⋅ K_mat → medium
* Q_sat ⋅ K_the → low

---

## Step 3: Use Values (final output)

Now we don’t take Keys anymore.

👉 We take **Values** based on importance:

```id="o6z9xp"
Output_sat = 
  (high weight × V_cat) +
  (medium weight × V_mat) +
  (low weight × others)
```

---

# 🔹 Why 3 different vectors?

Because each has a different role:

### 1. Query (Q)

👉 “What am I searching for?”

* For "sat" → maybe looking for subject (who sat?)

---

### 2. Key (K)

👉 “What do I represent?”

* "cat" → represents a subject → matches well

---

### 3. Value (V)

👉 “What information should I give if selected?”

* This is the actual content that flows forward

---

# 🔹 Super simple analogy

Think of a **classroom discussion**:

* You ask a question → **Query**
* Students raise hands based on relevance → **Keys**
* The answers they give → **Values**

👉 You listen more to relevant students

---

# 🔹 In one compact flow

```id="m5y8kl"
1. Q compares with all K → gives scores
2. Scores → converted to weights
3. Weights × V → final output
```

---

# 🔹 Important insight

👉 Q and K are used only for **matching**
👉 V is used for **actual information transfer**

---

# 🔹 Final takeaway (must remember)

👉

* Q = what I need
* K = what others offer
* V = what they actually give

👉 Self-attention =
**match (Q,K) → use V**

---

If you want next:
I can show you:
✅ exact formula (QKᵀ / √d) in simple way
✅ small numeric example (very useful for interviews)
✅ how multi-head attention builds on this

Just tell me 👍


# Multi-Head Attention

<div class="text_component" data-testid="text-component"><p style="text-align: justify;"><strong>Multi-Head Attention</strong></p><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">In the previous segment, you saw how the attention model calculates the context vector. However, the original paper suggests the usage of eight such attention heads, which it calls multi-head attention. In this segment, you will understand what exactly it is and how it is different from self-attention.&nbsp;</p><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">Rather than computing single attention (the weighted sum of values), multi-head attention computes multiple attention weighted sums (hence the name).</p><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">In the next video, Ankush will explain the need for multi-head attention.</p></div>

<div class="MuiBox-root css-1bi8ut6"><div class="text_component" data-testid="text-component"><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">As explained in the video, there are two major reasons for using multi-head attention:<br>If we use the 512 dimensions to calculate the context vector, the matrix multiplication between the query, key and value will be of a large order; therefore, the processing will be slower.</p><p style="text-align: justify;">&nbsp;</p><p style="text-align: center;">&nbsp;</p><p><img class="image-center" style="cursor: pointer; border: 2px solid transparent; display: block; margin-left: auto; margin-right: auto; text-align: center;" src="https://images.upgrad.com/2a9e5cb3-b00e-4836-bf77-7355bee0defd-Multi-Head Attention.png"></p><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">The input encodings are passed to eight different self-attention blocks, each of a dimension S∗dk, where&nbsp;dk = 64.&nbsp; Using multiple heads gives you a rich understanding of the same input sequences, as the context vector is calculated in multiple ways.</p></div></div>

<div class="MuiBox-root css-1bi8ut6"><div class="text_component" data-testid="text-component"><p style="text-align: justify;">Each ‘multiple-head’ is a linear transformation of the input representation. Using multiple linear transformations, &nbsp;you can extract different features of the input. But how do we calculate the final context vector resulting from the eight context vectors? You will understand this in the next video</p></div></div>

<div class="MuiBox-root css-1bi8ut6"><div class="text_component" data-testid="text-component"><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">The original paper suggests the usage of h = 8 (context vectors/attention heads); however, this is a hyperparameter that can be customised. As explained in the video, we will concatenate the context vectors from all eight heads. This concatenated matrix is multiplied with a weights matrix such that the output dimension is S x d_model(same as the dimension of input embeddings fed to the encoder). The weight matrix also helps to add in all the learnt linear projections from each of the context vectors.</p><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;"><b><b><span class="equation-content" data-latex="MultiHead(output)  = concat ( z_{1}, z_{2},z_{3},z_{4},z_{5},z_{6},z_{7})W^{0} = (S,d_{model})" data-widget="equation-editor"><span class="mjx-chtml"><span class="mjx-math" aria-label="MultiHead(output)  = concat ( z_{1}, z_{2},z_{3},z_{4},z_{5},z_{6},z_{7})W^{0} = (S,d_{model})"><span class="mjx-mrow" aria-hidden="true"><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.446em; padding-bottom: 0.298em; padding-right: 0.081em;">M</span></span><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.298em;">u</span></span><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.446em; padding-bottom: 0.298em;">l</span></span><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.372em; padding-bottom: 0.298em;">t</span></span><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.446em; padding-bottom: 0.298em;">i</span></span><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.446em; padding-bottom: 0.298em; padding-right: 0.057em;">H</span></span><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.298em;">e</span></span><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.298em;">a</span></span><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.446em; padding-bottom: 0.298em; padding-right: 0.003em;">d</span></span><span class="mjx-mo"><span class="mjx-char MJXc-TeX-main-R" style="padding-top: 0.446em; padding-bottom: 0.593em;">(</span></span><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.298em;">o</span></span><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.298em;">u</span></span><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.372em; padding-bottom: 0.298em;">t</span></span><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.446em;">p</span></span><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.298em;">u</span></span><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.372em; padding-bottom: 0.298em;">t</span></span><span class="mjx-mo"><span class="mjx-char MJXc-TeX-main-R" style="padding-top: 0.446em; padding-bottom: 0.593em;">)</span></span><span class="mjx-mo MJXc-space3"><span class="mjx-char MJXc-TeX-main-R" style="padding-top: 0.077em; padding-bottom: 0.298em;">=</span></span><span class="mjx-mi MJXc-space3"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.298em;">c</span></span><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.298em;">o</span></span><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.298em;">n</span></span><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.298em;">c</span></span><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.298em;">a</span></span><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.372em; padding-bottom: 0.298em;">t</span></span><span class="mjx-mo"><span class="mjx-char MJXc-TeX-main-R" style="padding-top: 0.446em; padding-bottom: 0.593em;">(</span></span><span class="mjx-msubsup"><span class="mjx-base" style="margin-right: -0.003em;"><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.298em; padding-right: 0.003em;">z</span></span></span><span class="mjx-sub" style="font-size: 70.7%; vertical-align: -0.212em; padding-right: 0.071em;"><span class="mjx-texatom" style=""><span class="mjx-mrow"><span class="mjx-mn"><span class="mjx-char MJXc-TeX-main-R" style="padding-top: 0.372em; padding-bottom: 0.372em;">1</span></span></span></span></span></span><span class="mjx-mo"><span class="mjx-char MJXc-TeX-main-R" style="margin-top: -0.144em; padding-bottom: 0.519em;">,</span></span><span class="mjx-msubsup MJXc-space1"><span class="mjx-base" style="margin-right: -0.003em;"><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.298em; padding-right: 0.003em;">z</span></span></span><span class="mjx-sub" style="font-size: 70.7%; vertical-align: -0.212em; padding-right: 0.071em;"><span class="mjx-texatom" style=""><span class="mjx-mrow"><span class="mjx-mn"><span class="mjx-char MJXc-TeX-main-R" style="padding-top: 0.372em; padding-bottom: 0.372em;">2</span></span></span></span></span></span><span class="mjx-mo"><span class="mjx-char MJXc-TeX-main-R" style="margin-top: -0.144em; padding-bottom: 0.519em;">,</span></span><span class="mjx-msubsup MJXc-space1"><span class="mjx-base" style="margin-right: -0.003em;"><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.298em; padding-right: 0.003em;">z</span></span></span><span class="mjx-sub" style="font-size: 70.7%; vertical-align: -0.212em; padding-right: 0.071em;"><span class="mjx-texatom" style=""><span class="mjx-mrow"><span class="mjx-mn"><span class="mjx-char MJXc-TeX-main-R" style="padding-top: 0.372em; padding-bottom: 0.372em;">3</span></span></span></span></span></span><span class="mjx-mo"><span class="mjx-char MJXc-TeX-main-R" style="margin-top: -0.144em; padding-bottom: 0.519em;">,</span></span><span class="mjx-msubsup MJXc-space1"><span class="mjx-base" style="margin-right: -0.003em;"><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.298em; padding-right: 0.003em;">z</span></span></span><span class="mjx-sub" style="font-size: 70.7%; vertical-align: -0.212em; padding-right: 0.071em;"><span class="mjx-texatom" style=""><span class="mjx-mrow"><span class="mjx-mn"><span class="mjx-char MJXc-TeX-main-R" style="padding-top: 0.372em; padding-bottom: 0.372em;">4</span></span></span></span></span></span><span class="mjx-mo"><span class="mjx-char MJXc-TeX-main-R" style="margin-top: -0.144em; padding-bottom: 0.519em;">,</span></span><span class="mjx-msubsup MJXc-space1"><span class="mjx-base" style="margin-right: -0.003em;"><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.298em; padding-right: 0.003em;">z</span></span></span><span class="mjx-sub" style="font-size: 70.7%; vertical-align: -0.212em; padding-right: 0.071em;"><span class="mjx-texatom" style=""><span class="mjx-mrow"><span class="mjx-mn"><span class="mjx-char MJXc-TeX-main-R" style="padding-top: 0.372em; padding-bottom: 0.372em;">5</span></span></span></span></span></span><span class="mjx-mo"><span class="mjx-char MJXc-TeX-main-R" style="margin-top: -0.144em; padding-bottom: 0.519em;">,</span></span><span class="mjx-msubsup MJXc-space1"><span class="mjx-base" style="margin-right: -0.003em;"><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.298em; padding-right: 0.003em;">z</span></span></span><span class="mjx-sub" style="font-size: 70.7%; vertical-align: -0.212em; padding-right: 0.071em;"><span class="mjx-texatom" style=""><span class="mjx-mrow"><span class="mjx-mn"><span class="mjx-char MJXc-TeX-main-R" style="padding-top: 0.372em; padding-bottom: 0.372em;">6</span></span></span></span></span></span><span class="mjx-mo"><span class="mjx-char MJXc-TeX-main-R" style="margin-top: -0.144em; padding-bottom: 0.519em;">,</span></span><span class="mjx-msubsup MJXc-space1"><span class="mjx-base" style="margin-right: -0.003em;"><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.298em; padding-right: 0.003em;">z</span></span></span><span class="mjx-sub" style="font-size: 70.7%; vertical-align: -0.212em; padding-right: 0.071em;"><span class="mjx-texatom" style=""><span class="mjx-mrow"><span class="mjx-mn"><span class="mjx-char MJXc-TeX-main-R" style="padding-top: 0.372em; padding-bottom: 0.372em;">7</span></span></span></span></span></span><span class="mjx-mo"><span class="mjx-char MJXc-TeX-main-R" style="padding-top: 0.446em; padding-bottom: 0.593em;">)</span></span><span class="mjx-msubsup"><span class="mjx-base" style="margin-right: -0.104em;"><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.446em; padding-bottom: 0.298em; padding-right: 0.104em;">W</span></span></span><span class="mjx-sup" style="font-size: 70.7%; vertical-align: 0.513em; padding-left: 0.262em; padding-right: 0.071em;"><span class="mjx-texatom" style=""><span class="mjx-mrow"><span class="mjx-mn"><span class="mjx-char MJXc-TeX-main-R" style="padding-top: 0.372em; padding-bottom: 0.372em;">0</span></span></span></span></span></span><span class="mjx-mo MJXc-space3"><span class="mjx-char MJXc-TeX-main-R" style="padding-top: 0.077em; padding-bottom: 0.298em;">=</span></span><span class="mjx-mo MJXc-space3"><span class="mjx-char MJXc-TeX-main-R" style="padding-top: 0.446em; padding-bottom: 0.593em;">(</span></span><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.519em; padding-bottom: 0.298em; padding-right: 0.032em;">S</span></span><span class="mjx-mo"><span class="mjx-char MJXc-TeX-main-R" style="margin-top: -0.144em; padding-bottom: 0.519em;">,</span></span><span class="mjx-msubsup MJXc-space1"><span class="mjx-base" style="margin-right: -0.003em;"><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.446em; padding-bottom: 0.298em; padding-right: 0.003em;">d</span></span></span><span class="mjx-sub" style="font-size: 70.7%; vertical-align: -0.219em; padding-right: 0.071em;"><span class="mjx-texatom" style=""><span class="mjx-mrow"><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.298em;">m</span></span><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.298em;">o</span></span><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.446em; padding-bottom: 0.298em; padding-right: 0.003em;">d</span></span><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.298em;">e</span></span><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.446em; padding-bottom: 0.298em;">l</span></span></span></span></span></span><span class="mjx-mo"><span class="mjx-char MJXc-TeX-main-R" style="padding-top: 0.446em; padding-bottom: 0.593em;">)</span></span></span></span></span></span></b></b></p><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">Let's quickly revise the entire process:</p><ul><li style="text-align: justify;">Calculate Query(Q), Key(K) and Value() by multiplying the input embeddings with their respective matrices&nbsp;<span class="equation-content" data-latex="W_{q},W_{k},W_{v}" data-widget="equation-editor"><span class="mjx-chtml"><span class="mjx-math" aria-label="W_{q},W_{k},W_{v}"><span class="mjx-mrow" aria-hidden="true"><span class="mjx-msubsup"><span class="mjx-base" style="margin-right: -0.104em;"><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.446em; padding-bottom: 0.298em; padding-right: 0.104em;">W</span></span></span><span class="mjx-sub" style="font-size: 70.7%; vertical-align: -0.212em; padding-right: 0.071em;"><span class="mjx-texatom" style=""><span class="mjx-mrow"><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.446em; padding-right: 0.014em;">q</span></span></span></span></span></span><span class="mjx-mo"><span class="mjx-char MJXc-TeX-main-R" style="margin-top: -0.144em; padding-bottom: 0.519em;">,</span></span><span class="mjx-msubsup MJXc-space1"><span class="mjx-base" style="margin-right: -0.104em;"><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.446em; padding-bottom: 0.298em; padding-right: 0.104em;">W</span></span></span><span class="mjx-sub" style="font-size: 70.7%; vertical-align: -0.219em; padding-right: 0.071em;"><span class="mjx-texatom" style=""><span class="mjx-mrow"><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.446em; padding-bottom: 0.298em;">k</span></span></span></span></span></span><span class="mjx-mo"><span class="mjx-char MJXc-TeX-main-R" style="margin-top: -0.144em; padding-bottom: 0.519em;">,</span></span><span class="mjx-msubsup MJXc-space1"><span class="mjx-base" style="margin-right: -0.104em;"><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.446em; padding-bottom: 0.298em; padding-right: 0.104em;">W</span></span></span><span class="mjx-sub" style="font-size: 70.7%; vertical-align: -0.212em; padding-right: 0.071em;"><span class="mjx-texatom" style=""><span class="mjx-mrow"><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.298em;">v</span></span></span></span></span></span></span></span></span></span></li><li style="text-align: justify;">Calculate the context vector (<span class="equation-content" data-latex="z" data-widget="equation-editor"><span class="mjx-chtml"><span class="mjx-math" aria-label="z"><span class="mjx-mrow" aria-hidden="true"><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.298em; padding-right: 0.003em;">z</span></span></span></span></span></span>) from each attention head.</li><li style="text-align: justify;">Concatenate all the context vectors&nbsp;<span class="equation-content" data-latex=" ( z_{1}, z_{2},z_{3},z_{4},z_{5},z_{6},z_{7})" data-widget="equation-editor"><span class="mjx-chtml"><span class="mjx-math" aria-label=" ( z_{1}, z_{2},z_{3},z_{4},z_{5},z_{6},z_{7})"><span class="mjx-mrow" aria-hidden="true"><span class="mjx-mo"><span class="mjx-char MJXc-TeX-main-R" style="padding-top: 0.446em; padding-bottom: 0.593em;">(</span></span><span class="mjx-msubsup"><span class="mjx-base" style="margin-right: -0.003em;"><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.298em; padding-right: 0.003em;">z</span></span></span><span class="mjx-sub" style="font-size: 70.7%; vertical-align: -0.212em; padding-right: 0.071em;"><span class="mjx-texatom" style=""><span class="mjx-mrow"><span class="mjx-mn"><span class="mjx-char MJXc-TeX-main-R" style="padding-top: 0.372em; padding-bottom: 0.372em;">1</span></span></span></span></span></span><span class="mjx-mo"><span class="mjx-char MJXc-TeX-main-R" style="margin-top: -0.144em; padding-bottom: 0.519em;">,</span></span><span class="mjx-msubsup MJXc-space1"><span class="mjx-base" style="margin-right: -0.003em;"><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.298em; padding-right: 0.003em;">z</span></span></span><span class="mjx-sub" style="font-size: 70.7%; vertical-align: -0.212em; padding-right: 0.071em;"><span class="mjx-texatom" style=""><span class="mjx-mrow"><span class="mjx-mn"><span class="mjx-char MJXc-TeX-main-R" style="padding-top: 0.372em; padding-bottom: 0.372em;">2</span></span></span></span></span></span><span class="mjx-mo"><span class="mjx-char MJXc-TeX-main-R" style="margin-top: -0.144em; padding-bottom: 0.519em;">,</span></span><span class="mjx-msubsup MJXc-space1"><span class="mjx-base" style="margin-right: -0.003em;"><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.298em; padding-right: 0.003em;">z</span></span></span><span class="mjx-sub" style="font-size: 70.7%; vertical-align: -0.212em; padding-right: 0.071em;"><span class="mjx-texatom" style=""><span class="mjx-mrow"><span class="mjx-mn"><span class="mjx-char MJXc-TeX-main-R" style="padding-top: 0.372em; padding-bottom: 0.372em;">3</span></span></span></span></span></span><span class="mjx-mo"><span class="mjx-char MJXc-TeX-main-R" style="margin-top: -0.144em; padding-bottom: 0.519em;">,</span></span><span class="mjx-msubsup MJXc-space1"><span class="mjx-base" style="margin-right: -0.003em;"><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.298em; padding-right: 0.003em;">z</span></span></span><span class="mjx-sub" style="font-size: 70.7%; vertical-align: -0.212em; padding-right: 0.071em;"><span class="mjx-texatom" style=""><span class="mjx-mrow"><span class="mjx-mn"><span class="mjx-char MJXc-TeX-main-R" style="padding-top: 0.372em; padding-bottom: 0.372em;">4</span></span></span></span></span></span><span class="mjx-mo"><span class="mjx-char MJXc-TeX-main-R" style="margin-top: -0.144em; padding-bottom: 0.519em;">,</span></span><span class="mjx-msubsup MJXc-space1"><span class="mjx-base" style="margin-right: -0.003em;"><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.298em; padding-right: 0.003em;">z</span></span></span><span class="mjx-sub" style="font-size: 70.7%; vertical-align: -0.212em; padding-right: 0.071em;"><span class="mjx-texatom" style=""><span class="mjx-mrow"><span class="mjx-mn"><span class="mjx-char MJXc-TeX-main-R" style="padding-top: 0.372em; padding-bottom: 0.372em;">5</span></span></span></span></span></span><span class="mjx-mo"><span class="mjx-char MJXc-TeX-main-R" style="margin-top: -0.144em; padding-bottom: 0.519em;">,</span></span><span class="mjx-msubsup MJXc-space1"><span class="mjx-base" style="margin-right: -0.003em;"><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.298em; padding-right: 0.003em;">z</span></span></span><span class="mjx-sub" style="font-size: 70.7%; vertical-align: -0.212em; padding-right: 0.071em;"><span class="mjx-texatom" style=""><span class="mjx-mrow"><span class="mjx-mn"><span class="mjx-char MJXc-TeX-main-R" style="padding-top: 0.372em; padding-bottom: 0.372em;">6</span></span></span></span></span></span><span class="mjx-mo"><span class="mjx-char MJXc-TeX-main-R" style="margin-top: -0.144em; padding-bottom: 0.519em;">,</span></span><span class="mjx-msubsup MJXc-space1"><span class="mjx-base" style="margin-right: -0.003em;"><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.298em; padding-right: 0.003em;">z</span></span></span><span class="mjx-sub" style="font-size: 70.7%; vertical-align: -0.212em; padding-right: 0.071em;"><span class="mjx-texatom" style=""><span class="mjx-mrow"><span class="mjx-mn"><span class="mjx-char MJXc-TeX-main-R" style="padding-top: 0.372em; padding-bottom: 0.372em;">7</span></span></span></span></span></span><span class="mjx-mo"><span class="mjx-char MJXc-TeX-main-R" style="padding-top: 0.446em; padding-bottom: 0.593em;">)</span></span></span></span></span></span> and multiply with a weights matrix such that the output&nbsp;(<span class="equation-content" data-latex="Z" data-widget="equation-editor"><span class="mjx-chtml"><span class="mjx-math" aria-label="Z"><span class="mjx-mrow" aria-hidden="true"><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.446em; padding-bottom: 0.298em; padding-right: 0.04em;">Z</span></span></span></span></span></span>) =&nbsp;<span class="equation-content" data-latex="(S,d_{model})" data-widget="equation-editor"><span class="mjx-chtml"><span class="mjx-math" aria-label="(S,d_{model})"><span class="mjx-mrow" aria-hidden="true"><span class="mjx-mo"><span class="mjx-char MJXc-TeX-main-R" style="padding-top: 0.446em; padding-bottom: 0.593em;">(</span></span><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.519em; padding-bottom: 0.298em; padding-right: 0.032em;">S</span></span><span class="mjx-mo"><span class="mjx-char MJXc-TeX-main-R" style="margin-top: -0.144em; padding-bottom: 0.519em;">,</span></span><span class="mjx-msubsup MJXc-space1"><span class="mjx-base" style="margin-right: -0.003em;"><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.446em; padding-bottom: 0.298em; padding-right: 0.003em;">d</span></span></span><span class="mjx-sub" style="font-size: 70.7%; vertical-align: -0.219em; padding-right: 0.071em;"><span class="mjx-texatom" style=""><span class="mjx-mrow"><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.298em;">m</span></span><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.298em;">o</span></span><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.446em; padding-bottom: 0.298em; padding-right: 0.003em;">d</span></span><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.225em; padding-bottom: 0.298em;">e</span></span><span class="mjx-mi"><span class="mjx-char MJXc-TeX-math-I" style="padding-top: 0.446em; padding-bottom: 0.298em;">l</span></span></span></span></span></span><span class="mjx-mo"><span class="mjx-char MJXc-TeX-main-R" style="padding-top: 0.446em; padding-bottom: 0.593em;">)</span></span></span></span></span></span></li></ul><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">&nbsp;</p><p style="text-align: center;">&nbsp;</p><p><img data-width="2755" data-height="1320" height="287.47731397459165" width="600" class="image-center" style="cursor: pointer; border: 2px solid transparent; display: block; margin-left: auto; margin-right: auto; text-align: center;" src="https://images.upgrad.com/7c721951-9164-4928-b2a1-c5808539b72a-attention_heads.png"></p><p style="text-align: justify;">When we encode the word "it", the first attention head captures only one feature of the given input sequences. But if we activate all 8 attention heads, the model captures multiple features of the same input sequence.</p><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">Now that you have understood the working of a multi-head attention block, let’s take a look at the next sub-layer of the encoder block: pointwise feed-forward network.</p><p style="text-align: justify;">&nbsp;</p><h2 style="text-align: justify;">Additional Readings</h2><p style="text-align: justify;"><strong>Visualizing Attention in Transformer</strong></p><p style="text-align: justify;">Check out this <a target="_blank" href="https://www.arxiv-vanity.com/papers/1904.02679/">link </a>to visualise different attention heads in Transformer architecture</p><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">Before you proceed to the next segment, please attempt the question given below.</p><p>&nbsp;</p></div></div>

Great — now that you understand **Q, K, V**, let’s make **Multi-Head Attention** feel *intuitive* 👇

---

# 🔹 One-line idea

👉 **Multi-head attention = multiple self-attentions running in parallel, each focusing on different things**

---

# 🔹 Why do we need it?

Single self-attention:

* Looks at relationships in **one way only**

👉 But language is complex:

* Grammar (subject, verb)
* Meaning (context)
* Relationships (who did what)

💡 One attention is not enough → so we use **multiple heads**

---

# 🔹 Simple analogy

Imagine reading this sentence:

👉 "The bank will not approve the loan"

Now you analyze it like:

* 👨‍🏫 Head 1 → looks for **grammar** (subject-verb)
* 💰 Head 2 → looks for **financial meaning**
* 🔗 Head 3 → looks for **word relationships**

👉 Each head focuses on a **different perspective**

---

# 🔹 How it works (step-by-step, simple)

### Step 1: Create multiple Q, K, V

Instead of one:

```id="p1"
Q, K, V
```

We create many:

```id="p2"
Q₁, K₁, V₁
Q₂, K₂, V₂
Q₃, K₃, V₃
...
```

Each is learned differently

---

### Step 2: Each head does self-attention

Each head independently:

```id="p3"
Head₁ = Attention(Q₁, K₁, V₁)
Head₂ = Attention(Q₂, K₂, V₂)
...
```

👉 Each head captures **different relationships**

---

### Step 3: Combine all heads

We then:

```id="p4"
Concatenate all heads → pass through a linear layer
```

👉 Final output = combined understanding

---

# 🔹 Intuition with example

Sentence:
👉 "The dog chased the ball because it was fast"

Different heads might learn:

* Head 1 → "dog → chased"
* Head 2 → "ball → fast"
* Head 3 → resolve "it" → ball

👉 Together → deep understanding

---

# 🔹 Visual intuition (mental model)

```id="p5"
Same sentence
   ↓
Split into multiple views
   ↓
Each head learns different patterns
   ↓
Combine all insights
```

---

# 🔹 Why it’s powerful

✅ Captures multiple relationships at once
✅ Learns richer context
✅ Handles ambiguity better
✅ Improves performance significantly

---

# 🔹 Important insight

👉 Each head has:

* its own Q, K, V matrices
* its own “attention style”

👉 So heads don’t repeat — they specialize

---

👉

* Single attention = one perspective
* Multi-head = multiple perspectives

But, sometimes we might lose positional information while moving from one operation to the next. <div class="MuiBox-root css-1bi8ut6"><div class="text_component" data-testid="text-component"><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">As you saw in the previous video, there is an ‘add and normalise’ layer after the ‘multi-head attention’ layer. The ‘residual connections’ added around the layers are the most important part of this sub-layer for the following reasons:</p><ul><li style="text-align: justify;">It helps retain the position-related information we add to the input representation/embedding across the network.&nbsp;</li><li style="text-align: justify;">Therefore, it allows us to look back at the input of the previous layer and its output simultaneously.</li><li style="text-align: justify;">Like the ResNet architecture, it also helps in avoiding gradient vanishing/explosion while backpropagating the gradients.</li></ul><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">Also, it has been observed that the network showed catastrophic results once the residual connections were removed.</p><p style="text-align: justify;">To regulate the computation, a normalisation operation is performed on the resulting output from the skip connection so each feature (column) has the same average and deviation.&nbsp;</p><p style="text-align: justify;"><br>The output from this block is 𝒙 ′ = LayerNorm(Multi-Head Attention(𝑿) + 𝑿).</p><p style="text-align: justify;"><br>The next layer is a position-wise fully connected feed-forward network. This network is applied to each position individually and exactly in the same way. It is a couple of linear layers with a ReLU activation function in between. Comparatively, it is identical to a two-layer convolution with a kernel size 1.&nbsp;</p></div></div>

# Understanding the Decoder

<div class="MuiBox-root css-1bi8ut6"><div class="text_component" data-testid="text-component"><p style="text-align: justify;"></p><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">The main function of the decoder block is to predict the next token of the sequence using the previous tokens and attention from the encoder block. While training the decoder, we pass the target output as the input to the decoder. However, it is shifted to the right through the addition of a &lt;start&gt; token, as no tokens were previously generated. Hence, using the &lt;start&gt; token and the context vector, the decoder predicts the first token.</p><p style="text-align: justify;"><br>Every word predicted subsequently is added as an input to the decoder in the next timestep and the whole process is repeated until the &lt;end&gt; token is predicted. Let’s understand this process in the next video.</p></div></div>



<div class="MuiBox-root css-1bi8ut6"><div class="text_component" data-testid="text-component"><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">As explained in the video, the encoder and the decoder work like the typical encoder-decoder architecture you saw in the previous module.&nbsp;</p><p style="text-align: center;">&nbsp;</p><p><img alt="Decoder block in the transformer architecture" class="image-center" style="cursor: pointer; border: 2px solid transparent; display: block; margin-left: auto; margin-right: auto; text-align: center;" src="https://images.upgrad.com/49d9a3b9-00fc-46ff-bdbf-0dd1939af858-combined.png"></p><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">&nbsp;</p></div></div>

<div class="MuiBox-root css-1bi8ut6"><div class="text_component" data-testid="text-component"><p style="text-align: justify;">In the diagram above of the decoder block, the output embedding and positional embedding layers have the same role and structure as in the encoder. However, the decoder block in the transformer architecture has some additional components. You will understand them in the next video.</p></div></div>

<div class="MuiBox-root css-1bi8ut6"><div class="text_component" data-testid="text-component"><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">As seen in the video, there are mainly two additions in the decoder block:</p><ul><li style="text-align: justify;">Cross-attention (encoder-decoder attention)</li><li style="text-align: justify;">Look ahead mask</li></ul><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">&nbsp;</p><h2 style="text-align: justify;">Look Ahead Mask</h2><p style="text-align: justify;"><br>The multi-head attention in the decoder is applied to tokens at the current timestep (index till which prediction is done by transformer) and not for the future tokens (the ones that are not predicted till now). However, since the input to the decoder is fed in parallel, we must block the future tokens. This can be done using a ‘look-ahead mask’, which is a weight matrix consisting of ‘-inf’ values in the upper diagonal and ‘0’ values in the lower diagonal. Therefore, the attention block with the look-ahead mask is referred to as the ‘masked multi-head attention’ layer.</p></div></div>



<div class="MuiBox-root css-1bi8ut6"><div class="text_component" data-testid="text-component"><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">&nbsp;Once the attention layers are applied to the current positions, a normalisation layer is applied, similarly to what we have done in the encoder.</p><h2 style="text-align: justify;"><br>Cross-Attention (Encoder-Decoder Attention)</h2><p style="text-align: justify;"><br>After applying the normalisation layer, a cross-attention (encoder-decoder attention) is applied to the output received from the previous layer and the output of the encoder.<br>This type of attention obtains its queries (Q) from the previous decoder layer (T*d_model, where T is the sequence length of the target), whereas the keys (K) and values (V) are acquired from the encoder output (S*d_model, where S is the sequence length of the source). This allows every position in the decoder to look over all the positions in the input sequence (similar to the typical encoder-decoder architecture).</p><p style="text-align: justify;">&nbsp;</p><p style="text-align: justify;">While this attention layer leverages pretrained values for the query, key, and value matrices, it still requires training to adjust these matrices to the specific task and data at hand.</p><p style="text-align: justify;"><img data-latex="CrossAttention = [output\_decoder_{sublayer-1}(Q), output\_encoder(K,V)]" alt="Equation" class="image-center" style="cursor: pointer; border: 2px solid transparent; display: block; margin-left: auto; margin-right: auto;" src="https://latex.upgrad.com/render?formula=CrossAttention%20%3D%20%5Boutput%5C_decoder_%7Bsublayer-1%7D(Q)%2C%20output%5C_encoder(K%2CV)%5D"></p><p style="text-align: justify;">&nbsp;</p><p>&nbsp;</p></div></div>
