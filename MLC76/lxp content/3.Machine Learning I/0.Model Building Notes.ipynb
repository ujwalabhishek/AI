### For categorical variables with multiple levels, create dummy features

* If your binary categorical variable is already encoded as 0 and 1,
 you do NOT need to create dummy variables.

 **Why?**

A binary variable (e.g., "Yes" or "No", "Male" or "Female") when encoded as 0 and 1 is already in numerical form and is ready for use in most machine learning models.

** Dummy variables are needed when:**

The feature has **more than two** categories (e.g., "Red", "Green", "Blue").

Or the feature is non-numeric and needs conversion to numeric.

| Original Feature                 | Encoded | Dummy Needed?            |
| -------------------------------- | ------- | ------------------------ |
| `Yes` / `No`                     | 1 / 0   |  No                     |
| `Male` / `Female`                | 1 / 0   |  No                     |
| `Color` (`Red`, `Blue`, `Green`) | Strings |  Yes, create dummy vars |
