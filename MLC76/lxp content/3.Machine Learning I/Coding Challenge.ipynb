### 1. Create another feature based called BMI_group which groups people based on their BMI. The groups should be as follows:
* Underweight: BMI is less than 18.5.
* Normal: BMI is 18.5 to 24.9.
* Overweight: BMI is 25 to 29.9.
* Obese: BMI is 30 or more.

The grouping is based on [WHO standards](http://www.euro.who.int/en/health-topics/disease-prevention/nutrition/a-healthy-lifestyle/body-mass-index-bmi).

The output should have first five rows of the resulting dataframe.

```python
import pandas as pd 
df=pd.read_csv("data/insurance.csv")
df.head()
```

```python
df['BMI_group'] = df['bmi'].apply(
    lambda x: 'Underweight' if x<18.5 else 
        'normal'if x<= 24.9  else
            'Overweight' if x<=29.9  else
                'Obese'
)

df.head()
```
